#[REF]https://ssd.jpl.nasa.gov/horizons.cgi
#[Target]
10: Sun 199: Mercury 299: Venus 399: Earth 499: Mars 599: Jupiter
699: Saturn 799: Uranus 899: Neptune 999: Pluto
#[Observer]
"500" Earth's center (same as "geocentric")
"@sun" center of the Sun
"@0" solar system barycenter (same as "@ssb")
"Viking 1@499" Viking 1 landing site on Mars (body 499)
"@hst" Hubble Space Telescope (same as "@-48")
"@-32" Voyager 2 spacecraft
"@301" center of the Moon
"Apollo 11@301" Apollo 11 landing site on the Moon's surface
"@phobos" center of Phobos, the Martian moon (same as "@401")

```
from astroquery.jplhorizons import Horizons
obj = Horizons(id='499',location='@phobos',epochs={'start':'2000-10-01', 'stop':'2100-10-10','step':'5d'})
vec = obj.vectors()
x = vec['x']
y = vec['y']
z = vec['z']
```

```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
g = plt.subplot()
#g.scatter(x, y, c=cm.jet(np.abs(z)), edgecolor='none')
g.scatter(x, y, c=cm.Spectral(np.abs(z)), edgecolor='none')
g.set_xlabel('X')
g.set_ylabel('Y')
g.set_title('Title')
g.set_aspect('equal')
plt.savefig('image.png')
plt.show()
```
